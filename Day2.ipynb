Create Arrays

```python
import numpy as np
a = np.array([1, 2, 3, 4, 5])
b = np.arange(1, 6)
c = np.zeros((3))
d = np.ones((3, 2))

print("Array a:",a)
print("Array b:",b)
print("2*3 Zeros:",c)
print("3*2 Ones:",d)
```

Indexing & Slicing

```python
print("First element of a:",a[0])
print("Last two  element of a:",a[-2:])
```

Basic arithmetic operations

```python
print("a + 2:", a + 2)
print("a * 3:", a * 3)
```

Basic Statistics

```python
print("square of a:", np.square(a))
print("mean of a:", np.mean(a))
print("standard deviation:", np.std(a))
```

Reshaping arrays

```python
matrix = np.arange(1, 10).reshape((3, 3))
print("Reshaped 3*3 matrix")
matrix
```

Transpose

```python
print("Transpose")
matrix.T
```

Boolean indexing

```python
print("elements > 5:",matrix[matrix > 5])
```

Pandas:

Create a DataFrame manually

```python
import pandas as pd
school = {
    'Name':['Sabrin','Sneha','Aishu','Altu'],
    'U-No':[10 ,20, 30, 40,],
    'Fees':[1000, 2000,30000,50000],
    'Department':['BCA','BBA','BA','B-Com']

}
df = pd.DataFrame(school)
df
```

Basic operation

```python
print("first 2 rows:")
df.head(2)
```

```python
print("colomn names:")
df.columns
```

```python
print("data types:")
df.dtypes
```

```python
print("describ numerical columns:")
df.describe()
```

Filtering

```python
print("people with fees > 2000:")
df[df['Fees'] > 2000]
```

Sorting

```python
print("sorted by fees:")
df.sort_values(by='Fees')
```

Grouping and Aggregation

```python
grouped = df.groupby('Department')['Fees'].mean()
print("Average Fees by department:")
grouped
```

Adding new columns

```python
df['Fees in Lakha'] = df['Fees'] /100000
df
```

Matplotlib:

Lien Plot

```python
import matplotlib.pyplot as plt

years = [2000, 2005, 2010, 2015, 2020]
income = [10, 20, 30, 40, 50]

plt.plot(years, income, marker='*', linestyle=':', color='red')
plt.title("Company income")
plt.xlabel("Years")
plt.ylabel("income(in Crores)")
plt.xticks(years)
plt.show()
```

Bar chart

```python
departments = df['Department'].value_counts()
departments.plot(kind='bar', color='orange')
plt.title("Employee Count by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

Histogram

```python
fees = df['Fees']
plt.hist(fees, bins=6, color='pink', edgecolor='black')
plt.title("fees Distribution")
plt.xlabel("Fees")
plt.ylabel("Frequency")
plt.show()
```

Pie Chart

```python
df.groupby('Department').size().plot(kind='pie', autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```
